```python


# Convertion Of Text
def extract_genres(text):
    # Converting String Into list
    actual_list = ast.literal_eval(text)
    
    # Keeping only The Name from Every Element
    names_only = [item['name'] for item in actual_list]
    
    # It Joins them in a text Separated with Commas
    return ", ".join(names_only)

# 2. Application In The Function with.apply()
movies_merged["genres"] = movies_merged["genres"].apply(extract_genres)
```

```python
# Dataset Overview
movies_merged["genres"].head()
```

```python
# Conversion to 0,1
movies_encoded = pd.get_dummies(movies_merged, columns=["genres"], dtype=int)
```

```python
# Data Overview
movies_encoded.shape
```

```python
movies_encoded.columns
```

```python
# Clearly Seperation 
genres_encoded = movies_merged["genres"].str.get_dummies(sep=", ")

# We Connect Them to Our Original dataset
movies_final = pd.concat([movies_merged, genres_encoded], axis=1)
```

```python
# Dataset Overview
movies_final.shape
```

```python
# 42. Dataset Overview
movies_final.columns
```

```python
# .43 Implementing Basic Features
basic_features = ["budget","runtime","release_year"]

# .44 Characteristics That Are Created From Genres
genres_features = genres_encoded.columns.tolist()

# .45 Combining The Two Lists
all_features = basic_features + genres_features


```

```python
# 46 Setting Up X and y
X = movies_final[all_features]
y = movies_final["is_successful"]
```
